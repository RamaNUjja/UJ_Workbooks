```python
'''
Question 4 -
1. Take a string as input from the user.
2. Check whether that string is a palindrome.
   Create a function which returns "PALINDROME" if true otherwise "NOT A PALINDROME."
'''
def lower_case(s):
  lower=""
  for i in s:
    if i >= 'A' and i <= 'Z':
      lower += chr(ord(i)+32)
    else:
      lower += i
  return lower

def palindrome(s):
  u=lower_case(s)
  return "PALINDROME" if u == u[::-1] else "NOT A PALINDROME."
  # Use of Ternary Operators

palindrome('Nitin')
```

```python
'''
Question 5 -
1. Take a string as input from the user.
2. Print the count of uppercase characters in that string. (alphabets)

INPUT - "This is a Random%^&*()"
OUTPUT - 2
'''
def count(s):
  count = 0
  for i in s:
    if i >= 'A' and i <= 'Z':
      count += 1
  return count

count(input())
```
