```python
def replace_cube(arr, N):
  res_list =[]
  i=0
  while i < len(arr):
    if arr[i]%N == 0:
      res_list.append((arr[i])**3)
    else:
      res_list.append(arr[i])
    i+=1
  return res_list

X_bar= [1,5,12,6,8,9,3]
replace_cube(X_bar, 9)
```

```python
def replace_cube(arr, N):
  res_list =[]
  for i in range(len(arr)):
    if (i+1)%N==0:
      res_list.append((arr[i])**3)
    else:
      res_list.append(arr[i])
  return res_list

X_bar= [15,18,20,14,20,13,2,19,8,12]
replace_cube(X_bar, 3)
```
