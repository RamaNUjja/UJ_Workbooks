```python
def show2(w,w0,x,y):
  x_val= np.linspace(-5,15,50)
  y_val= -(w[0]*x_val + w0)/w[1]
  plt.plot(x_val, y_val)
  plt.scatter(x[:,0],x[:,1], c= y) # x[:,0] is x coordiante of point & x[:,1] is y coordinate of point
  plt.xlim(x[:,0].min()-1.5, x[:,0].max()+1.5)
  plt.ylim(x[:,1].min()-1.5,x[:,1].max()+1.5)
  plt.show()
```

```python
show2(w,w0,x,y)
```

```python
def get_prediction(w,w0,xi):
  return (np.dot(w,xi)+w0)/np.linalg.norm(w)
```

```python
def training(x,y,w,w0,num_iter = 100):
  print("INITIAL STATE")
  show2(w,w0,x,y)
  for i in range(1, num_iter+1):
    print(f"Iteration:{i}")
    mcpc=0
    total_points= len(y)
    for xi, yi in zip(x,y):
      predection= np.sign(get_prediction(w,w0,xi))
      if predection!=np.sign(yi):
        mcpc += 1
        # updation of line perform
        w = w + (xi*yi)
        w0 = w0 + np.sign(yi)
    print(f"Accuracy: {(total_points-mcpc)*100/ total_points}%")
    show2(w,w0,x,y)
    if mcpc==0:# all is correct
      return w, w0
  return w, w0

training(x,y,w,w0,num_iter= 100)
```

```python
def partial_dev(func, x, y):
  delta=0.0001
  # partial dev of func wrt x and y
  fx= (func(x+delta,y)-func(x,y))/delta
  fy= (func(x,y+delta)-func(x,y))/delta
  return fx, fy
```

```python
# This function takes derivative of f at point x
def derivative(x, f):
 delta = 0.0001
 return (f(x + delta) - f(x)) / delta
# This function does exactly one update to my parameter x
def one_step_gd(x, f, lr): # lr -> learning rate, same as eta(n)
 grad = derivative(x, f)
 new_x = x - lr * grad
 new_fx = f(new_x) # new y
 return new_x, new_fx, grad
```

```python
## Define the initial random guess
x0 = -5
num_steps = 0 # Same as t in the update equation
# Let us save all our steps.
xs = [] # all the x's
fxs = [] # all the y's
grads = []
```

```python
## Lets plot!!
plt.figure()
# First, plot the function
x = np.arange(-10, 10, 0.01)
y = f(x)
plt.plot(x, y)
# Now apply gradient descent
new_x, new_fx, grad = one_step_gd(x0, f, 0.1)
xs.append(new_x)
fxs.append(new_fx)
grads.append(grad)
num_steps += 1
# Next, plot the old guess and the gradient at that point
t_c = f(x0) - grads[-1] * x0
delta = 2.
t_x = np.arange(x0 - delta, x0 + delta, 0.01)
t_y = grads[-1] * t_x + t_c
plt.plot(t_x, t_y, color="black")
# Next, plot the other guesses
for i in range(num_steps-1):
  plt.plot(xs[i], fxs[i], marker="o", markersize=5, markeredgecolor="green", markerfacecolor="green")
plt.plot(x0, f(x0), marker="o", markersize=5, markeredgecolor="red", markerfacecolor="red")
plt.plot(xs[-1], fxs[-1], marker="o", markersize=5, markeredgecolor="orange", markerfacecolor="orange")
print(new_x)
x0 = new_x
plt.show()
```

```python
## Implement PCA!!
# Step 1: Column standardization

mu = data.mean(axis = 0)

std = data.std(axis = 0)

X_st = (data - mu)/std

# Step 2: Eigenvalue decomposition
cov_X_st = np.matmul(X_st.values.T, X_st.values)
eigenvalues, eigenvectors = np.linalg.eig(cov_X_st)

print("eigenvalues:",eigenvalues, sep="\n")
print("eigenvectors:",eigenvectors, sep="\n")

# Step 3: Reduce the dimension! Choose the eigenvector corresponding to the largest eigenvalue
max_idx = np.argmax(eigenvalues)
X_prime = eigenvectors[:, max_idx].dot(X_st.T)
X_prime_reduced = X_prime

# Step 4: Visualize
plt.figure()
plt.scatter(X_prime_reduced, np.zeros(20))
plt.show()
```
