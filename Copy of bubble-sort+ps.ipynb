```python
def bubble_sort1(a):

  n = len(a)

  #ith iteration over the list
  for i in range(n):

    for j in range(0,n):

      if a[j] > a[j+1]:
        #swap both
        a[j] , a[j+1] = a[j+1], a[j]

  return a
```

```python
a = [5,1,2,4,7,3]
ans = bubble_sort1(a)
```

```python
def bubble_sort1(a):

  n = len(a)

  #ith iteration over the list
  for i in range(n):

    for j in range(0,n-1):

      if a[j] > a[j+1]:
        #swap both
        a[j] , a[j+1] = a[j+1], a[j]

  return a
```

```python
a = [5,1,2,4,7,3]
ans = bubble_sort1(a)
```

```python
print(ans)
```

```python
def bubble_sort2(a):

  n = len(a)

  #ith iteration over the list
  for i in range(n):

    for j in range(0,n-1-i):

      if a[j] > a[j+1]:
        #swap both
        a[j] , a[j+1] = a[j+1], a[j]

  return a
```

```python
a = [5,1,2,4,7,3]
ans = bubble_sort2(a)
print(ans)
```

```python
#Level2: If given a already sorted list: [1, 2, 3, 4, 5, 7]
```

```python
def bubble_sort3(a):

  n = len(a)

  #ith iteration over the list
  for i in range(n-1):

    is_sorted = True

    for j in range(0,n-1-i):

      if a[j] > a[j+1]:
        #swap both
        a[j] , a[j+1] = a[j+1], a[j]
        is_sorted = False

    #no more iterations required in the outer loop
    if is_sorted == True:
      break

  return a
```

```python
#TC: for already sorted list: O(n)
```

```python
#logic1

n = 475
sum = 0

while n > 0:
  last_digit = n%10
  sum += last_digit
  n = n//10

print(sum)
```
