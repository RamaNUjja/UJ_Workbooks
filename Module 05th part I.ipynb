```python
'''
Qustion 2 -
runs_virat = [67, 54, 12, 34, 77, 89, 101]

In runs_virat -> Calculate, sum, avg, max and min runs.
Without using any internal functions. You can use the length function.
'''
# Do not use sum, avg, min, max

runs_virat = [67, 54, 12, 34, 77, 89, 101]
sum=0
count=0
for i in runs_virat:
  sum += i
  count += 1
print('Total-',sum)
print('Avg-',(sum/count))

MAX=runs_virat[0]
for i in range(1, len(runs_virat)):
  if runs_virat[i]> MAX:
    MAX= runs_virat[i]
print('maximum-', MAX)

MIN= runs_virat[0]
for i in range(1, len(runs_virat)):
  if runs_virat[i]< MIN:
    MIN= runs_virat[i]
print('minimum-', MIN)
```

```python
#46/ Hw-1
def first_last(A,B):
  C=[]
  for i in range(len(A)):
    if A[i]==B:
      C.append(i)
  if len(C)==1:
    return C+C
  else:
    return C[0::(len(C)-1)]

first_last([1,5,6,4,7,12,23,7,0,2],4)
```

```python
#42/ HW-02
def star(n):
  for i in range(n):
    for j in range(n-i):
      if i>0 and i<(n-2) and j>0 and j<n-i-1:
        print(' ', end='')
      else:
        print('*', end='')
    print()
```

```python
#42/ HW-04
def perfect(N):
  sum=0
  for D in range(1,int(N/2)+1):
    if N%D==0:
      sum += D
  if sum==N:
    return 1
  else:
    return 0

perfect(6)
```

```python
#43/ As-01
def BMI(y,x):
  '''x is weight in kg and y is height in cm, it helps to get BMI of the person
  '''
  BMI= round((x/(y*y))*10000, 1)
  if BMI>29.9:
    print('Obese')
    return BMI
  elif 24.9<BMI<=29.9:
    print('Overweight')
    return BMI
  elif 18.5<=BMI<=24.9:
    print('Normal')
    return BMI
  elif BMI<18.5:
    print('Underweight')
    return BMI
BMI(171,51)
```

```python
#43/ As-03
def corona(A,B,C):
  ''' A is cases recovered per day, B is New cases per day and C is Current active case'''
  days=1
  i=1
  while i>=1:
    if (B+C)>A:
      days+=i
      C+=(B-A)
    else:
      break
  return days

corona(29,25,10000)
```
